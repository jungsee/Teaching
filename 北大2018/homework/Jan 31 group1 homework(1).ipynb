```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy import optimize 
```

```python
S,p,c = 20,0.5,0.5

def ϕ():
    return np.random.geometric(p) - 1



def G(x, d, s):
    if x <= s:
        return max(S - d, 0)
    else:
        return max(x - d, 0)

def π(x, d, s):
    if x <= s:
        return min(S, d) - c
    else:
        return min(x, d)

def each(X,n,s):
    for t in range(n-1):
        D = ϕ()
        X[t+1] = G(X[t], D, s)
        P[t] = π(X[t], D, s)
    return P.mean()

def A_π(X,n,s):
    y = np.empty(len(s))
    for i in range(len(s)):
        y[i] =  each(X,n,s[i])
    return y
```

```python
n = 100
X = np.empty(n)
for r in range(21):
    X[0] = r
    P = np.empty(n)
    AP = np.empty(21)
    fig, ax = plt.subplots(figsize=(7, 4.5))
    s = np.linspace(0,20,2000)
    y = A_π(X,n,s)
    ax.plot(s, y)
    ax.set(ylabel='$average profit$', xlabel='$s$')
    ax.set_yticks(np.linspace(0,2,5))
    print('The maximum of average profit can be achieved at s =  ',round(s[y.argmax()],2))
    print('The maximum of average profit is', y.max())
    plt.show()
```

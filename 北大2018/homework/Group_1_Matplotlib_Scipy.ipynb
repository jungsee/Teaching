## Matplotlib 

```python
# Plot  sin(x) and  cos(x) functions with  x∈[−π,π] and their crossing points
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()
x = np.linspace(-np.pi,np.pi,150)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x,y1,'b-',linewidth=2,alpha=0.6,label=r'$y1=\sin(x)$')
ax.plot(x,y2,'g-',linewidth=2,alpha=0.6,label=r'$y2=\cos(x)$')

t1 = 0.25 * np.pi
ax.scatter([t1, ], [np.cos(t1), ], 50, color='red')
ax.annotate(r'$cross point1$',
             xy=(t1, np.sin(t1)), xycoords='data',
             xytext=(+5, +15), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
t2 = -0.75 * np.pi
ax.scatter([t2, ], [np.cos(t2), ], 50, color='red')
ax.annotate(r'$cross point2$',
             xy=(t2, np.sin(t2)), xycoords='data',
             xytext=(+5, +15), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

ax.set_xticks([-0.75*np.pi,0.25*np.pi])
ax.set_yticks([np.sin(-0.75*np.pi),np.sin(0.25*np.pi)])
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
from scipy.stats import linregress

x = np.random.randn(20)
y = 2 * x + 0.5 * np.random.randn(20)
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_reg = slope * x + intercept   #the regression line

fig,ax = plt.subplots()
ax.scatter([x, ], [y, ], 50, color='blue')   # plot the samples
ax.plot(x,y_reg,'r-')  # plot the regression line

plt.show()
```

### Exercise 2

```python
from scipy.optimize import fsolve

def f(x):
    x0, x1 = x.tolist()
    return [
        x0**2 - 3*x0 + 7 -385,
        (x0+x1)/2-1.5
    ]

r = fsolve(f, [1, 1])
print(r)
```

```python
def bisec(f,a,b,len):
    t = (a+b)/2
    if f(a)*f(b)>0:
            return 'No root exists in this interval.'
    while b-a>len:
        if f(t)== 0:
            return t
        elif f(t)*f(a)>0:
            return bisec(f,t,b,len)  # substitute a with t
        else:
            return bisec(f,a,t,len)  # substitute b with t
    return t
```
